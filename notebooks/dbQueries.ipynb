```python
from neo4j import GraphDatabase

# Adjust these to match your local Neo4j setup
NEO4J_URI = "neo4j://127.0.0.1:7687"
NEO4J_USER = "neo4j"
NEO4J_PASSWORD = "graphrag"  

driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))

# Test connection
with driver.session() as session:
    result = session.run("RETURN 1 AS test")
    print("Connected to Neo4j:", result.single()["test"] == 1)
```

```python
#GRAPH SCHEMA STATISTICS
-- Node counts by label
MATCH (n) RETURN labels(n), count(n)

-- Relationship counts by type  
MATCH ()-[r]->() RETURN type(r), count(r)

-- Average degree (density)
MATCH (n)-[r]-() RETURN avg(count(r))

-- Most connected entities
MATCH (n)-[r]-() RETURN n.name, labels(n), count(r) ORDER BY count(r) DESC LIMIT 20
```
